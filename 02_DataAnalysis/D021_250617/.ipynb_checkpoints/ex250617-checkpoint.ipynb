# 데이터 분석 기초

**데이터 분석의 순서**  
정제 -> 분석 -> 시각화

**데이터 정제 기법**  
- 변환: 다양한 형태의 값들을 일관된 형태로 바꾸는 작업.
- 파싱: 일정 규칙 하에 유의미한 단위로 분할하는 작업.
- 보강: 변환, 파싱, 수정, 표준화 등을 통한 추가정보를 작업.

````python
import pandas as pd
path = '../data/' 
````

count: 빈도(frequency) - 값의 개수  
mean: 평균  
std: 표준편차(standard deviation) - 변수의 값들이 평균에서 떨어진 정도  
min: 최소값(minimum)  
25%: 1사분위수(1st quantile) - 하위 25% 지점에 위치한 값  
50%: 중앙값(median) - 하위 50% 지점에 위치한 값  
75%: 3사분위수(3rd quantile) - 하위 75% 지점에 위치한 값  
max: 최대값(maximum)

## 파생변수 만들기

기존 변수를 변형해 만든 변수를 파생변수라고 한다.  

````python
df = pd.DataFrame({'var1' : [ 4, 3, 8 ], 'var2' : [ 2, 6, 1 ]})
df
````

````python
# var_sum 파생변수 만들기
df['var_sum'] = df['var1'] + df['var2']
df
````

````python
# var_mean 파생변수 만들기
df['var_mean'] = (df['var1'] + df['var2']) / 2
df
````

### 조건문을 활용해 파생변수 만들기

**1. 기준값 정하기**  

````python
import matplotlib.pyplot as plt
````

````python
mpg['total'].describe()
````

````python
# 그래프를 만드는 코드
mpg['total'].plot.hist()
# 그래프를 그리는 코드
plt.show()
````

히스토그램을 통해 연비가 20이 넘는 차에 고연비 합격 판정을 내리면 될 것 같다.  

**2. 합격 판정 변수 만들기**

````python
import numpy as np
````

Numpy는 N차원 데이터를 다룰 때 사용되며, Pandas와 마찬가지로 C언어로 구현되었기 때문에 연산속도가 빠르다.

Numpy의 Where를 사용하여 데이터를 필터링 할 수 있다.
```
np.where(조건, 조건이 맞을 때, 조건이 맞지 않을 때)
```

````python
# 20 이상이면 pass, 그렇지 않으면 fail
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')
mpg.head()
````

**3. 빈도표로 합격 판정 자동차 수 살펴보기**

````python
# 연비 합격 빈도표 만들기
mpg['test'].value_counts()
````

**4. 막대 그래프로 빈도 표현하기**  

````python
# 연비 합격 빈도표를 변수에 할당
count_test = mpg['test'].value_counts()
count_test
````

````python
# 연비 합격 빈도 막대 그래프 만들기
count_test.plot.bar()
````

**축 이름 회전하기**

````python
# 축 이름 수평으로 만들기
count_test.plot.bar(rot=0)
````

히스토그램은 연속적 데이터를 표현할 때 유용하다.  
x축은 구간값, y는 빈도이다.  
막대간 간경이 없고, 붙어있다.  

바 그래프는 각 개별 항목(요소)에 대한 개수 값을 표현한다.  
각 바가 분리되어 있다.  

### 중첩 조건문

**1. 연비 등급 변수 만들기**

````python
# total 기준으로 A, B, C 등급 부여
mpg['grade'] = np.where(mpg['total'] >= 30, 'A', np.where(mpg['total'] >= 20, 'B', 'C'))

mpg.head()
````

**2. 빈도표와 막대 그래프로 연비 등급 살펴보기**

````python
# 등급 빈도표 만들기
count_grade = mpg['grade'].value_counts()
count_grade
````

````python
# 등급 빈도 막대 그래프 만들기
count_grade.plot.bar(rot=0)
````

````python
# 등급 빈도표 알파벳순 정렬
count_grade = mpg['grade'].value_counts().sort_index()
count_grade
````

````python
# 등급 빈도 막대 그래프 만들기
count_grade.plot.bar(rot=0)
````

### 목록에 해당하는 행으로 변수 만들기

````python
# |(vertical bar)를 사용하면 조건문의 or을 사용할 수 있다.
mpg['size'] = np.where((mpg['category'] == 'compact') |
                       (mpg['category'] == 'subcompact') |
                       (mpg['category'] == '2seater'), 'small', 'large')

mpg['size'].value_counts()
````

````python
# isin()으로 여러 조건문 사용하기
mpg['size'] = np.where(mpg['category'].isin(['compact', 'subcompact', '2seater']), 'small', 'large')

mpg['size'].value_counts()
````

문제3. `total`, `asian` 변수를 이용해 '전체 인구 대비 아시아 인구 백분율' 파생변수를 추가하고, 히스토그램을 만들어 분포를 살펴보세요

````python
new_midwest['asian_per_total'].plot.hist()
````

문제4. 아시아 인구 백분율 전체 평균을 구하고, 평균을 초과하면 `'large'`, 그 외에는 `'small'`을 부여한 파생변수를 만들어 보세요

### lambda 이용해 데이터 프레임 명 줄여 쓰기

요약 통계량

- 평균 mean()
- 표준편차 std()
- 합계 sum()
- 중앙값 median()
- 최소값 min()
- 최대값 max()
- 빈도(개수) count()

위 나열된 통계 함수들을 그룹으로 나누어 적용 => groupby()

## 분석 도전 문제

Q1. `popadults`는 해당 지역의 성인 인구, `poptotal`은 전체 인구를 나타냅니다. `midwest` 데이터에 '전체 인구 대비 미성년 인구 백분율' 변수를 추가하세요.

````python
midwest = pd.read_csv('../data/midwest.csv')
midwest
````

````python
midwest = midwest.assign(minors_per_total = lambda x: (x['poptotal'] - x['popadults']) / x['poptotal'] * 100)
midwest['minors_per_total']
````

Q2. 미성년 인구 백분율이 가장 높은 상위 5개 `county`(지역)의 미성년 인구 백분율을 출력하세요.

````python
midwest.groupby('county').agg(minors_max = ('minors_per_total', 'max')).sort_values('minors_max', ascending=False).head()
````

Q3. 분류표의 기준에 따라 미성년 비율 등급 변수를 추가하고, 각 등급에 몇 개의 지역이 있는지 알아보세요.

````python
midwest = midwest.assign(minors_scale = np.where(midwest['minors_per_total'] >= 40, 'large', 
                                        np.where(midwest['minors_per_total'] >= 30, 'middle', 'small')))
midwest.groupby('minors_scale').agg(county = ('county', 'count'))
````

Q4. `popasian`은 해당 지역의 아시아인 인구를 나타냅니다. '전체 인구 대비 아시아인 인구 백분율' 변수를 추가하고 하위 10개 지역의 `state`(주), `county`(지역), 아시아인 인구 백분율을 출력하세요.

````python
midwest.assign(asian_per_total = lambda x: x['popasian'] / x['poptotal'] * 100)[['state', 'county', 'asian_per_total']].sort_values('asian_per_total').head(10)
````

## 이상치(이상한 값) 정제

### 이상치 제거

````python
# 1, 2로 구분되는 성별 데이터에 이상치 3이 들어가 있다.
df = pd.DataFrame({'sex' : [1, 2, 1, 3, 2, 1], 'score' : [5, 4, 3, 4, 2, 6]})
df
````

**이상치 확인**

````python
df['sex'].value_counts().sort_index()
````

````python
df['score'].value_counts().sort_index()
````

**이상치를 결측 처리**

````python
# sex가 3이면 NaN 부여
df['sex'] = np.where(df['sex'] == 3, np.nan, df['sex'])
df
````

````python
# score가 5보다 크면 NaN 부여
df['score'] = np.where(df['score'] > 5, np.nan, df['score'])
df
````

````python
df.dropna(subset=['sex', 'score']).groupby('sex').agg(mean_score = ('score', 'mean'))
````

np.where()는 반환하는 값 중에 문자가 있으면 NaN이 아닌 'nan'을 반환한다. 그래서 NaN이 아닌 다른 문자열로 결측치를 처리하고, 나중에 바꿔주면 된다.

### 극단적인 값

논리적으로 존재할 수 있지만 극단적으로 크거나 작은 값을 극단치(outlier)이라고 한다.  

**상자 그림으로 극단치 기준 정하기**

Q1(1사분위수) ~ Q3(3사분위수)의 구간을 IQR(사분위 범위, Q3 - Q1의 50%인구간)이라고 한다. IQR의 가운데가 Q2(Median) 중앙값이다.  
Q1 - 1.5xIQR ~ Q3 + 1.5xIQR 구간의 바깥쪽에 있는 값을 극단치라고 판단할 수 있다.

**1. 상자 그림 살펴보기**

````python
import seaborn as sns

mpg = pd.read_csv(path + 'mpg.csv')
sns.boxplot(data=mpg, y='hwy')
````

**2. 극단치 기준값 구하기**

````python
# 1사분위수
pct25 = mpg['hwy'].quantile(.25)
pct25
````

````python
# 3사분위수
pct75 = mpg['hwy'].quantile(.75)
pct75
````

````python
# IQR
iqr = pct75 - pct25
iqr
````

````python
# 하한
pct25 - 1.5 * iqr
````

````python
# 상한
pct75 + 1.5 * iqr
````

**3. 극단치를 결측 처리하기**

````python
# 4.5 ~ 40.5 벗어나면 NaN 부여
mpg['hwy'] = np.where((mpg['hwy'] < 4.5) | (mpg['hwy'] > 40.5), np.nan, mpg['hwy'])

# 결측치 빈도 확인
mpg['hwy'].isna().sum()
````

**4. 결측치 제거하고 분석하기**

````python
mpg.dropna(subset = 'hwy').groupby('drv').agg(mean_hwy=('hwy', 'mean'))
````
