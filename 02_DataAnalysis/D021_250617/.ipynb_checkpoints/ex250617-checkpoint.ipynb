# 데이터 분석 기초

**데이터 분석의 순서**  
정제 -> 분석 -> 시각화

**데이터 정제 기법**  
- 변환: 다양한 형태의 값들을 일관된 형태로 바꾸는 작업.
- 파싱: 일정 규칙 하에 유의미한 단위로 분할하는 작업.
- 보강: 변환, 파싱, 수정, 표준화 등을 통한 추가정보를 작업.

````python
import pandas as pd
path = '../data/' 
````

count: 빈도(frequency) - 값의 개수  
mean: 평균  
std: 표준편차(standard deviation) - 변수의 값들이 평균에서 떨어진 정도  
min: 최소값(minimum)  
25%: 1사분위수(1st quantile) - 하위 25% 지점에 위치한 값  
50%: 중앙값(median) - 하위 50% 지점에 위치한 값  
75%: 3사분위수(3rd quantile) - 하위 75% 지점에 위치한 값  
max: 최대값(maximum)

## 파생변수 만들기

기존 변수를 변형해 만든 변수를 파생변수라고 한다.  

````python
df = pd.DataFrame({'var1' : [ 4, 3, 8 ], 'var2' : [ 2, 6, 1 ]})
df
````

````python
# var_sum 파생변수 만들기
df['var_sum'] = df['var1'] + df['var2']
df
````

````python
# var_mean 파생변수 만들기
df['var_mean'] = (df['var1'] + df['var2']) / 2
df
````

### 조건문을 활용해 파생변수 만들기

**1. 기준값 정하기**  

````python
import matplotlib.pyplot as plt
````

````python
mpg['total'].describe()
````

````python
# 그래프를 만드는 코드
mpg['total'].plot.hist()
# 그래프를 그리는 코드
plt.show()
````

히스토그램을 통해 연비가 20이 넘는 차에 고연비 합격 판정을 내리면 될 것 같다.  

**2. 합격 판정 변수 만들기**

````python
import numpy as np
````

Numpy는 N차원 데이터를 다룰 때 사용되며, Pandas와 마찬가지로 C언어로 구현되었기 때문에 연산속도가 빠르다.

Numpy의 Where를 사용하여 데이터를 필터링 할 수 있다.
```
np.where(조건, 조건이 맞을 때, 조건이 맞지 않을 때)
```

````python
# 20 이상이면 pass, 그렇지 않으면 fail
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')
mpg.head()
````

**3. 빈도표로 합격 판정 자동차 수 살펴보기**

````python
# 연비 합격 빈도표 만들기
mpg['test'].value_counts()
````

**4. 막대 그래프로 빈도 표현하기**  

````python
# 연비 합격 빈도표를 변수에 할당
count_test = mpg['test'].value_counts()
count_test
````

````python
# 연비 합격 빈도 막대 그래프 만들기
count_test.plot.bar()
````

**축 이름 회전하기**

````python
# 축 이름 수평으로 만들기
count_test.plot.bar(rot=0)
````

히스토그램은 연속적 데이터를 표현할 때 유용하다.  
x축은 구간값, y는 빈도이다.  
막대간 간경이 없고, 붙어있다.  

바 그래프는 각 개별 항목(요소)에 대한 개수 값을 표현한다.  
각 바가 분리되어 있다.  

### 중첩 조건문

**1. 연비 등급 변수 만들기**

````python
# total 기준으로 A, B, C 등급 부여
mpg['grade'] = np.where(mpg['total'] >= 30, 'A', np.where(mpg['total'] >= 20, 'B', 'C'))

mpg.head()
````

**2. 빈도표와 막대 그래프로 연비 등급 살펴보기**

````python
# 등급 빈도표 만들기
count_grade = mpg['grade'].value_counts()
count_grade
````

````python
# 등급 빈도 막대 그래프 만들기
count_grade.plot.bar(rot=0)
````

````python
# 등급 빈도표 알파벳순 정렬
count_grade = mpg['grade'].value_counts().sort_index()
count_grade
````

````python
# 등급 빈도 막대 그래프 만들기
count_grade.plot.bar(rot=0)
````

### 목록에 해당하는 행으로 변수 만들기

````python
# |(vertical bar)를 사용하면 조건문의 or을 사용할 수 있다.
mpg['size'] = np.where((mpg['category'] == 'compact') |
                       (mpg['category'] == 'subcompact') |
                       (mpg['category'] == '2seater'), 'small', 'large')

mpg['size'].value_counts()
````

````python
# isin()으로 여러 조건문 사용하기
mpg['size'] = np.where(mpg['category'].isin(['compact', 'subcompact', '2seater']), 'small', 'large')

mpg['size'].value_counts()
````

문제2. `poptotal`(전체 인구) 변수를 `total`로, `popasian`(아시아 인구) 변수를 `asian`으로 수정하세요

문제3. `total`, `asian` 변수를 이용해 '전체 인구 대비 아시아 인구 백분율' 파생변수를 추가하고, 히스토그램을 만들어 분포를 살펴보세요

````python
new_midwest['asian_per_total'].plot.hist()
````

문제4. 아시아 인구 백분율 전체 평균을 구하고, 평균을 초과하면 `'large'`, 그 외에는 `'small'`을 부여한 파생변수를 만들어 보세요

### lambda 이용해 데이터 프레임 명 줄여 쓰기

````python
# 평균 연비 변수 추가
mpg['avg_total'] = (mpg['cty'] + mpg['hwy']) / 2

# 평균 연비 변수 상위 3개 조회
mpg.sort_values('avg_total', ascending=False).head(3)
````

````python
# 위 코드들을 하나로 연결
mpg.assign(avg_total = lambda x: (x['cty'] + x['hwy']) / 2).sort_values('avg_total', ascending=False).head(3)
````

## 집단별로 요약하기

df.agg(): aggregate 집계 합수.  
요약 통계량  
요약 통계의 종류
- 평균 mean()
- 표준편차 std()
- 합계 sum()
- 중앙값 median()
- 최소값 min()
- 최대값 max()
- 빈도(개수) count()

위 나열된 통계 함수들을 그룹으로 나누어 적용 => groupby()

agg(속성명, 요약통계함수)

````python
# math 평균 구하기
exam.agg(mean_math = ('math', 'mean'))
````

### 집단별 요약 통계량 구하기

````python
exam.groupby('nclass').agg(mean_math = ('math', 'mean'))
````

nclass 변수가 인덱스로 들어간다.  
groupby(변수명, as_index=False)로 변수를 인덱스로 바꾸지 않게 할 수 있다.

### 집단별로 다시 집단 나누기

````python
# 제조회사로 그룹을 짓고, 제조회사별 구동 방식으로 그룹을 짓는다.
mpg.groupby(['manufacturer', 'drv']).agg(mean_cty = ('cty', 'mean'))
````

````python
# 특정 제조사
mpg.query('manufacturer == "audi"').groupby(['drv']).agg(n = ('drv', 'count'))
````

### pandas 함수 조합하기

Q. 제조회사별로 'suv' 자동차의 도시 및 고속도로 합산 연비 평균을 구해 내림차순으로 정렬하고, 1-5위까지 출력하기

````python
mpg.query('category == "suv"').assign(total = lambda x: (x['cty'] + x['hwy']) / 2).groupby('manufacturer').agg(total = ('total', 'mean')).sort_values('total', ascending=False).head(5)
````

### 혼자서 해보기 문제

Q1. mpg 데이터의 category는 자동차를 특징에 따라 suv, compact 등 일곱 종류로 분류한 변수입니다. 어떤 차종의 도시 연비가 높은지 비교해 보려고 합니다. category 별 cty 평균을 구해보세요.

````python
mpg = pd.read_csv('../data/mpg.csv')
mpg
````

````python
mpg.groupby('category').agg(avg_cty = ('cty', 'mean'))
````

Q2. 앞 문제의 출력 결과는 category 값 알파벳순으로 정렬되어 있습니다. 어떤 차종의 도시 연비가 높은지 쉽게 알아볼 수 있도록 cty 평균이 높은 순으로 정렬해 출력하세요.

````python
mpg.groupby('category').agg(avg_cty = ('cty', 'mean')).sort_values('avg_cty', ascending = False)
````

Q3. 어떤 회사 자동차의 hwy가 가장 높은지 알아보려고 합니다. hwy 평균이 가장 높은 회사 세 곳을 출력하세요.

````python
mpg.groupby('manufacturer').agg(max_hwy = ('hwy', 'max')).sort_values('max_hwy', ascending=False).head(3)
````

Q4. 어떤 회사에서 'compact' 차종을 가장 많이 생산하는지 알아보려고 합니다. 회사별 'compact' 차종 수를 내림차순으로 출력하세요.

````python
mpg.query('category == "compact"').groupby('manufacturer').agg(compact_cnt = ('category', 'count')).sort_values('compact_cnt', ascending=False)
````
