```python
import pandas as pd
import numpy as np

data = pd.read_csv('data.csv', encoding='cp949')
data
```

```python
data.isna().sum()
```

```python
data['합계 3시간 신적설(cm)'].unique()
```

```python
data['합계 3시간 신적설(cm)'] = data['합계 3시간 신적설(cm)'].fillna(0)
```

```python
data.isna().sum()
```

```python
data['일강수량(mm)'] = data['일강수량(mm)'].fillna(0)
```

```python
for col in data.columns:
    if type(data[col][0]) is str:
        continue

    data[col] = data[col].fillna(data[col].mean())
```

```python
data.isna().sum()
```
