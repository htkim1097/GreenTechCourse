```python
df_mid = df_mid.rename(columns={'poptotal' : 'total', 'popasian' : 'asian'})
```

```python
df_mid
```

```python
df_mid['total_per_asian'] = df_mid['asian'] / df_mid['total'] * 100
```

```python
df_mid
```

```python
df_mid['total_per_asian'].plot.hist()
```

```python
total_per_asian_mean = df_mid['total_per_asian'].mean()
df_mid['asian_amount'] = np.where(df_mid['total_per_asian'] > total_per_asian_mean, 'large', 'small')
df_mid['asian_amount']
```

```python
df_mid['asian_amount'].value_counts()
```

```python
df_mid['asian_amount'].value_counts().plot.bar()
```
